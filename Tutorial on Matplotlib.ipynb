```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Line graphs 

APPL_stock_price_7d = [10.2, 10.7, 9.9, 10.4, 9.6, 11.0, 11.2]

plt.plot(APPL_stock_price_7d, marker='o')

#add legends
plt.legend(['Apple'])

#customize x axis
plt.xticks(np.arange(7), ('Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'))

# add labels to x/y axis and set title
plt.xlabel('Days')
plt.ylabel('Prices in USD')
plt.title('Stock price for Apple')

plt.show()
```

```python
# Multiline graphs
# assuming we have stock prices for multiple organisations for last 7 days

apple = np.random.randint(low=6, high=15, size=7)
yahoo = np.random.randint(low=4, high=16, size=7)
google = np.random.randint(low=6, high=14, size=7)

#add markers to highlight datapoints
plt.plot(apple, marker='o')
plt.plot(yahoo, marker='*')
plt.plot(google, marker='x')

#customize x axis
plt.xticks(np.arange(7), ('Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'))

# add legend
plt.legend(['Apple', 'Yahoo', 'Google'])


# add labels to x/y axis and set title
plt.xlabel('Days of week')
plt.ylabel('Prices in USD')
plt.title('Stock price for Apple, Yahoo and Google')
plt.grid()
plt.show()
```

```python
# Bar graphs
# We shall use the USA 2016 election dataset

#load the dataset
data = pd.read_csv(r'C:\Users\rahul18.mishra\Desktop\Projects\matplotlib\data\2016_election_data.csv')

trump = data['Trump']

# graph number of polls conducted by each agency
survey = data.Pollster.value_counts().values
companies = data.Pollster.value_counts().keys()

#number of items on x axis
placeholder = np.arange(len(survey))

#make graph
plt.bar(placeholder, survey, align='center', alpha=.5)

#add name of the companies on x-axis
plt.xticks(placeholder, companies)

#add title and labels
plt.title('Number of polls conducted by each agencies')
plt.ylabel('Nnumber of polls')

plt.grid()
plt.show()

'''
OBSERVATIONS
1) We can see that MC conduscted the highest surveys whereas FOX news conducted the least surverys
'''
```

```python
#another way of plotting bar-graph directly from dataframe

data.groupby(['Pollster']).agg({'Trump' : np.mean, 'Clinton' : np.mean}).plot(kind='bar', 
                                                                              title='Popularity indes : Trump v/s Clinton')

obs ='''
OBSERVATIONS
1) We can see that in every polls Clinton's approval rating suprpassed Trump
'''

print(obs)
```

```python
#SCATTER plot - scatter plot is used to shows how one variable effects another
#here we shall see how Trumps rise affected Clinton

#data cleaning
popularity = data.sort_values('Entry Date/Time (ET)')[['Clinton', 'Trump', 'Entry Date/Time (ET)', 'Other']]
popularity['Year-Month'] = [x[0:7] for x in popularity['Entry Date/Time (ET)']]
popularity = popularity.groupby('Year-Month').agg({'Clinton' : np.mean, 'Trump' : np.mean, 'Other' : np.mean})
popularity['Diff'] = np.abs(popularity['Trump'] - popularity['Clinton']) * 20

placeholder = np.arange(popularity.shape[0])
popularity.head()
```

```python
plt.scatter(placeholder, popularity['Clinton'], c='r', alpha=.4, s=popularity['Diff'])
plt.scatter(placeholder, popularity['Trump'], c='b', alpha=.7, s=popularity['Diff'])
plt.xlabel('Months 1-15')
plt.ylabel('Approval rating')
plt.title('Popularity comparision between Trump and Clinton by time')
plt.legend()
plt.grid()
plt.show()

obs = '''
OBSERVATIONS : 

1)here we can clearly see that Trumps approval rating started growing from 3rd month onwards and in few cases 
it also surpassed Clinton

2) We can also conclude that in first 6 months, Clintons popularity kept sinking as Trumps popularity kept growing,
however both the candidates popularity ratings started to sink after 10th month (mostl likely due to presidential debates where they both 
exposed each others scams)

3) The size of the bubbles represent the difference of points between both candidates, Initially both red and blue bubbles are bigger in size
representing the huge difference of opinions of voters and then they keep shrinking in size
'''
print(obs)
```
